# График №1

```python
data = pd.read_csv('hotel_bookings.csv')

def adr_year(data):
    return round(data.adr.mean())
a = data.groupby('arrival_date_year').apply(adr_year)

a.plot(kind='bar')
plt.title('ADR from August 2015 to August 2017\n')
plt.xlabel('\nYear')
plt.ylabel('\nADR')
```

ОПИСАНИЕ ГРАФИКА:
Изменения средней стоимости гостиничного номера в 178 странах за период с августа 2015 г. по август 2017 г.

ВЫВОДЫ: Ежегодный рост средней стоимости номера (ADR).

# График №2

```python
data = pd.read_csv('hotel_bookings.csv')

data = data.groupby(['hotel', 'arrival_date_year'])['adr'].mean()

data.unstack().plot(kind='bar', rot=45, stacked=False)
plt.title('ADR by hotel type in 2015-2017\n')
plt.xlabel('Hotel')
plt.ylabel('ADR')
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
```

ВЫВОДЫ:

# График №3.

```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 10,7

data = pd.read_csv('hotel_bookings.csv')
top = 11

data_top = data.country.value_counts().head(top).reset_index() #рейтинг из top
data_top.rename(columns={'index': 'country', 'country': 'booking'}, inplace=True) #замена названий столбцов

other_countries = len(data.country.unique()) - top
sum_other = data.country.value_counts().sum() - data.country.value_counts().head(top).sum() #other

data_top = data_top.append(pd.DataFrame([[f'Other {other_countries} countries', sum_other]], columns = ['country', 'booking']), ignore_index=True)
data_top.set_index(['country'], inplace = True) 

data_top.plot(kind='pie', y='booking')
plt.title('Total reservations by 178 countries')
plt.ylabel('')
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
```

ОПИСАНИЕ ГРАФИКА:
Суммарное количество заявок на бронирование всех типов гостиниц за период с августа 2015 г. по август 2017 г. в разрезе 178 стран.

ВЫВОДЫ: 
За период 2015-2017 гг.:
1. более 2/3 заявок на бронирование гостиниц было в 1 из 178 стран;
2. наибольшее количество заявок на бронирование гостиниц было в Португалии;

# График 4.

```python
data = pd.read_csv('hotel_bookings.csv')

def number_month(row):
    '''Нумерация месяцев'''
    
    months = {'January': '01', 'February': '02', 'March': '03', 'April': '04', 'May': '05', 'June': '06',
          'July': '07', 'August': '08', 'September': '09', 'October': '10', 'November': '11', 'December': '12'}

    for month, number in months.items():
        if row['arrival_date_month'] in month:
            return number    

data['number_month'] = data.apply(number_month, axis=1)
data

data = data.groupby(['arrival_date_year', 'number_month'])['adr'].count().plot()
```

# График 6.

```python
data = pd.read_csv('hotel_bookings.csv')

def number_month(row):
    '''Нумерация месяцев'''
    
    months = {'January': '01', 'February': '02', 'March': '03', 'April': '04', 'May': '05', 'June': '06',
          'July': '07', 'August': '08', 'September': '09', 'October': '10', 'November': '11', 'December': '12'}

    for month, number in months.items():
        if row['arrival_date_month'] in month:
            return number    

data['number_month'] = data.apply(number_month, axis=1)

data = data.groupby(['arrival_date_year', 'number_month'])['adr'].count().reset_index()
sns.distplot(data.adr)
plt.xlabel('\nCount of bookings')
```

ОПИСАНИЕ ГРАФИКА:
Распределение количества заявок на бронирование всех типов гостиниц по месяцам с августа 2015 г. по август 2017 г.
