Задание 2. По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) 
категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.


```python
import pandas as pd
data = pd.read_csv('power.csv')

filtered_countries = data[ (data['country']=='Latvia') | (data['country']=='Lithuania') | (data['country']=='Estonia')]
new_data = filtered_countries.sort_values(by=['year', 'category' ], ascending=[True, True])

a = new_data[ (new_data.year >= 2005) & (new_data.year <= 2010) & 
            ((new_data.category == 4) | (new_data.category == 12) | (new_data.category == 21)) & 
            (new_data.quantity >= 0)]

print('Суммарное потребление товаров странами Прибалтики (Латвия, Литва и Эстония) по годам и группам товаров \nпредставлено в таблице ниже. ')
a.groupby(['year','category'])['quantity'].sum().reset_index()
```
