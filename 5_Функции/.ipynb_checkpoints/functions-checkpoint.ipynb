Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов и перечень полок, на которых находятся документы, 
хранятся в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006', '7676 76'],
    '3': []
}  


#РЕШЕНИЕ

user_command = input('Введите команду: ')     
#----------------------------------------------------------------------------------------------------------------------
if user_command == 'p':
    def person_name(employee):        
        """
        1. Запрашивает номер документа и возвращает имя человека, которому он принадлежит.
        """
        user_input_num = input('Введите номер документа: ')
        
        i = 0
        for num in employee:
            if num['number'] == user_input_num:
                print("\nДокумент принадлежит:", num['name'])
                i += 1
                
        if i == 0:
            print('\nВы ввели несуществующий документ. Повторите ввод.')

    person_name(documents) 
#----------------------------------------------------------------------------------------------------------------------
elif user_command == 's':
    def shelf_number(directories):
        """
        2. Запрашивает номер документа и возвращает номер полки, на которой он находится. 
        Выводит сообщение, если введен несуществующий документ.
        """        
        user_input_num = input('Введите номер документа: ')
        
        i = 0
        for num in directories:
            for num1 in directories[num]:

                if num1 == user_input_num:
                    print('\nДокумент хранится на полке №', num)                
                    i += 1             

        if i == 0:
            print('\nВы ввели несуществующий документ. Повторите ввод.')

    shelf_number(directories) 
#----------------------------------------------------------------------------------------------------------------------
elif user_command == 'l':
    def docs_list(employee):
        """
        3. Выводит список всех документов в  формате: "passport "2207 876234" "Василий Гупкин" 
        """
        for i in employee:        
            print(f'{i["type"]} "{i["number"]}" "{i["name"]}"')

    docs_list(documents)   
#----------------------------------------------------------------------------------------------------------------------
elif user_command == 'as':
    def add_number_new_shelf(shelfs):    
        """
        4. Запрашивает номер новой полки и добавляет ее в перечень
        """
        new_shelf = input('Введите номер новой полки: ')
        directories.setdefault(new_shelf,[])  #добавить пару ключ-значение, если такого ключа еще нет в словаре
        print(directories) 

    add_number_new_shelf(directories)
#----------------------------------------------------------------------------------------------------------------------
elif user_command == 'd':              
    def del_doc(documents, directories): 
        """
        5. Запрашивает номер документа и удаляет его из каталога и из перечня полок, на которой он находится. 
        Выводит сообщение, если введен несуществующий документ.
        """
        del_number = input('Введите номер документа для удаления: ')
        i = 0
        j = 0

        for row in documents:

            if del_number == row['number']:
                documents.remove(row)
                i += 1

        if i == 0:
            print('\nУдаление невозможно, документа в каталоге не существует.')       

        for row in directories.items():

            if del_number in row[1]:
                list_doc = row[1]
                list_doc.remove(del_number)
                directories.setdefault(row[0], list_doc)  
                j += 1

        if j == 0:
            print('\nУдаление невозможно, документа на полках не существует.')
                  
        if i != 0 and j != 0:
            print(documents)
            print(directories) 

    del_doc(documents, directories) 
#----------------------------------------------------------------------------------------------------------------------
elif user_command == 'm':
    def move_doc(directories):
        """
        6. Запрашивает номер документа и целевую полку, а затем перемещает его с текущей полки на целевую.
           Выводит сообщение, когда пользователь пытается переместить несуществующий документ или переместить документ 
           на несуществующую полку
        """
        doc_number = input('Введите номер документа: ')
        shelf = input('Введите номер целевой полки: ')      

        i = 0

        if shelf not in directories.keys():
            print('\nВведенной полки не существует')
            return 

        else:
            for row in directories.items():
                if doc_number in row[1]:
                    list_documents = row[1] # все документы на полке, на котором хранится документ для перемещения
                    list_documents.remove(doc_number) #удаление с полки документа для перемещения                     

                    i += 1                

            if i == 0:
                print('\nУказанного документа нет')
            else:
                s = (list(directories.keys()))[int(shelf)-1]
                n = (list(directories.values()))[int(shelf)-1]
                n.append(doc_number)  
                print('\nФинальные полки: ', directories)

    move_doc(directories)                
#----------------------------------------------------------------------------------------------------------------------
elif user_command == 'a':
    def add_doc(documents, directories):
        """
        7. Запрашивает номер и тип документа, имя владельца и номер полки, на котором он будет храниться. 
           Добавляет новый документ в каталог и в перечень полок. 
        """        
        doc_number = input('Ведите номер документа: ')
        doc_type = input('Ведите тип документа: ')
        person_name = input('Ведите имя владельца документа: ')  
        shelf_number = input('Ведите номер полки: ')                      
       
        new_data = [doc_type, doc_number, person_name]
        doc_content = list(documents[1])  
        new_person = dict(zip(doc_content, new_data))    
        documents.append(new_person)
                  
        if shelf_number in (list(directories.keys())):
            a = (list(directories.keys()))[int(shelf_number)-1]
            b = (list(directories.values()))[int(shelf_number)-1]
            b.append(doc_number)  
                  
        else:
            directories.setdefault(shelf_number,[]) #добавить пару ключ-значение, если такого ключа еще нет в словаре  
            a = (list(directories.keys()))[len(directories.keys()) - 1]
            b = (list(directories.values()))[len(directories.keys()) - 1]
            b.append(doc_number)  
    
        print('\n', documents,  '\n')
        print(directories)  

    add_doc(documents, directories)                    
```

# Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
1- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
2- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
Правильно обработайте ситуацию, когда пользователь будет вводить несуществующий документ.
3- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
4- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень
5- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. Предусмотрите сценарий, когда пользователь вводит несуществующий документ;  
6- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку;  
7- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.
